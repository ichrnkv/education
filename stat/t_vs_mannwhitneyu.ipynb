## T-test vs Mann-Whitney

```python
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt

# настройка размеров графиков
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 8)

# настройка стиля графиков
plt.style.use('fivethirtyeight')
```

```python
# сгенерирую 2 заведомо отличных респределения

x = np.random.exponential(0.7, size=1000)
y = np.random.binomial(1, 0.7, size=1000)

plt.hist(x, bins = 10, alpha=.5, label = 'x')
plt.hist(y, bins = 2, alpha=.5, label = 'y')
plt.title('Распределение x и y')
plt.legend();
```

Распределения, очевидно, отличаются и стат. критерий должен это показать.

### T-тест:

```python
print(stats.ttest_ind(x, y))
```

Видим, что T-тест не позволяет отклонить нулевую гипотезу.  

Выясним, почему сравнив средние значения распределений.

```python
print('x mean value is %.4f' %np.mean(x))
print('y mean value is %.4f' %np.mean(y))
```

Ага, средние распределений действительно очень близки друг к другу.  
Доверять такому результату нельзя, тк не было выполнено условие о нормальности распределений.

### Критерий Манна-Уитни:

```python
print(stats.mannwhitneyu(x, y))
```

Критерий Манна-Уитни показывает что выборки отличаются.
